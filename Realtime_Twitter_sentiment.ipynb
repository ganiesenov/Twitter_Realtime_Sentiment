```python
def preprocessing_text (text): 
    text = text.lower()
    text = re.sub("[^a-zA-Z#]", " ", text)
    text = re.sub("\s+", " ", text)
    text = re.sub("http\S+", '', text)
    return text 
clasifier = TextClassifier.load("en-sentiment")
def get_sentiment(tweet): 
    sentence = Sentence(tweet)
    clasifier.predict(sentence)
    return str(sentence.labels).split("\'")[3]
```

```python
while True: 
    tweets = api.search_recent_tweets(query).data
    
    for tweet in tweets:
        original_tweet = tweet.text
        clean_text = preprocessing_text(original_tweet)
        sentiment = get_sentiment(clean_text)
        print('---------------------------------tweet------------------------------')
        print(original_tweet)
        print('---------------------------------------------------------------------')
        print('Sentiment:', sentiment)
        time.sleep(1)
        print('\n\n')
```
